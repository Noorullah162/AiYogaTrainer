````python
import os
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import math

def load_data():
  datasets = '/content/datasets/Vehicle-Detection-1/'
  image_size = (300, 300)

  train_images = []
  train_labels = []

  test_images = []
  test_labels = []

  valid_images = []
  valid_labels = []
  images = []

  for folder in os.listdir(datasets):
    print(folder)
    if folder  == 'train' or folder  == 'test' or folder  == 'valid':
      for file in tqdm(os.listdir(os.path.join(datasets, folder + '/images/'))):
        if folder  == 'train':
          img_path = os.path.join(os.path.join(datasets, folder + '/images/'), file)
          image = cv2.imread(img_path)
          image = cv2.resize(image, image_size)
          # cv2_imshow(image)

          train_images.append(image)
          # print(img_path, 'completed')
          f = datasets+folder+'/labels/' +file[:-4:] + '.txt'
          my_file = open(f, "r")
          # reading the file
          data = my_file.read()

          # replacing end splitting the text
          # when newline ('\n') is seen.
          l = data.split("\n")
          # print(data_into_list)
          f = False
          for i in l:
            if i[0] == '1':
              # l1 = i.split()
              # print(l1, math.floor(float(l1[1])))
              # image = image[math.floor(float(l1[1])) : math.floor(float(l1[3])), math.floor(float(l1[2])) : math.floor(float(l1[4]))]
              # # image = cv2.imread(image)
              # print(image)
              # if len(image) != 0 :
              #   print("hi")
              #   cv2_imshow(image)
              f = True
              break
          if f == False:
            train_labels.append(0)
          else:
            train_labels.append(1)
          my_file.close()


        elif folder  == 'test':
          img_path = os.path.join(os.path.join(datasets, folder + '/images/'), file)
          image = cv2.imread(img_path)
          image = cv2.resize(image, image_size)
          # cv2_imshow(image)

          test_images.append(image)
          # print(img_path, 'completed')
          f = datasets+folder+'/labels/' +file[:-4:] + '.txt'
          my_file = open(f, "r")
          # reading the file
          data = my_file.read()

          # replacing end splitting the text
          # when newline ('\n') is seen.
          l = data.split("\n")
          # print(data_into_list)
          f = False
          for i in l:
            if i[0] == '1':
              f = True
              break
          if f == False:
            test_labels.append(0)
          else:
            test_labels.append(1)
          my_file.close()


        elif folder  == 'valid':
          img_path = os.path.join(os.path.join(datasets, folder + '/images/'), file)
          image = cv2.imread(img_path)
          image = cv2.resize(image, image_size)
          # cv2_imshow(image)

          valid_images.append(image)
          # print(img_path, 'completed')
          f = datasets+folder+'/labels/' +file[:-4:] + '.txt'
          my_file = open(f, "r")
          # reading the file
          data = my_file.read()

          # replacing end splitting the text
          # when newline ('\n') is seen.
          l = data.split("\n")
          # print(data_into_list)
          f = False
          for i in l:
            if i[0] == '1':
              f = True
              break
          if f == False:
            valid_labels.append(0)
          else:
            valid_labels.append(1)
          my_file.close()

  train_images = np.array(train_images, dtype = 'float32')
  train_labels = np.array(train_labels, dtype = 'int32')

  test_images = np.array(test_images, dtype = 'float32')
  test_labels = np.array(test_labels, dtype = 'int32')

  valid_images = np.array(valid_images, dtype = 'float32')
  valid_labels = np.array(valid_labels, dtype = 'int32')

  return train_images, train_labels, test_images, test_labels, valid_images, valid_labels

train_images, train_labels, test_images, test_labels, valid_images, valid_labels = load_data()
````

````python
print(train_labels)
````

````python
print(train_labels.shape)
````

````python
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, Dense, AveragePooling2D

model = Sequential()

model.add(Conv2D(filters = 6, kernel_size = (5, 5), activation = 'relu', input_shape = (300, 300, 3) ))

model.add(AveragePooling2D(pool_size = (2,2)))

model.add(Conv2D(filters = 16, kernel_size = (5, 5), activation = 'relu' ))

model.add(AveragePooling2D(pool_size = (2,2)))

model.add(Flatten())

model.add(Dense(units = 120, activation = 'relu'))

model.add(Dense(units = 84, activation = 'relu'))

model.add(Dense(units = 2, activation = 'sigmoid'))

````

````python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
````

````python
model.fit(train_images, train_labels, epochs=7, batch_size = 32, validation_data=(valid_images, valid_labels))
````

````python
results = model.evaluate(test_images, test_labels)
print("Loss of the model  is - test ", results[0])
print("Accuracy of the model is - test", results[1]*100, "%")
````

## **Applying on Video**

````python
from ultralytics.yolo.engine.model import YOLO

model = YOLO("/content/drive/MyDrive/Car object Detection/Caryolomodel.pt")
````

Note :
Result will be store in a path which is written in code


**eg :** cv2.VideoWriter( ***path*** , cv2.VideoWriter_fourcc(*'MP4V'), 25, size)

````python
import cv2
from google.colab.patches import cv2_imshow

#load video
video = cv2.VideoCapture('/content/drive/MyDrive/t6.mp4')


if (video.isOpened() == False):
	print("Error reading video file")

# We need to set resolutions.
# so, convert them from float to integer.
frame_width = int(video.get(3))
frame_height = int(video.get(4))

size = (frame_width, frame_height)

# result storing and saving in path
result_output = cv2.VideoWriter('/content/drive/MyDrive/Car object Detection/output.mp4',
						cv2.VideoWriter_fourcc(*'MP4V'),
						25, size)

while(True):
  ret, frame = video.read()
  if ret == True:
    results = model.predict(frame)
    result = results[0]
    box = result.boxes

    # print(type(box))
    pred_boxes = box.xyxy
    pred_labels = box.cls
    pred_scores = box.conf
    print(pred_labels)
    print(pred_boxes)

    img1 = frame

    for i in range(len(pred_boxes)):
      # Only display boxes with confidence > 0.92 and considering only car class label
      box = pred_boxes[i].tolist()
      label = pred_labels[i].item()
      score = pred_scores[i].item()
      if label == 1.0 and score > 0.92:

        font = cv2.FONT_HERSHEY_SIMPLEX
        score = score * 100
        s = "car" + " : " + "{:.2f}".format(score)
        # print(box)
        b1, b2, b3, b4 = int(box[0]), int(box[1]), int(box[2]), int(box[3]),

        cv2.rectangle( img1 , ( b1, b2 ), ( b3 , b4), (0,0,255), 1)

        cv2.putText(img1, s , (b1 , b2- 7), font, 0.7, (0,0,255) , 2)
        # plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], fill=False, edgecolor=color, linewidth=1)

        print(score)
    result_output.write(img1)
    if cv2.waitKey(1) & 0xFF == ord('s'):
      break
  # Break the loop
  else:
    break

# When everything done, release
# the video capture and video
# write objects
video.release()

# result saving
result_output.release()

# Closes all the frames
cv2.destroyAllWindows()

print("The video was successfully saved")
````
